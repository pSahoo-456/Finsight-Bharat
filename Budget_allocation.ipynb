```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

def load_data(file_path):
    df = pd.read_csv(file_path)
    return df

def plot_budget_trends(df):
    plt.figure(figsize=(12, 6))
    for column in df.columns[1:-1]:
        plt.plot(df['Year'], df[column], label=column)
    plt.xlabel("Year")
    plt.ylabel("Budget Allocation (Crores)")
    plt.title("Budget Allocation Trends (1990-2024)")
    plt.legend()
    plt.grid(True)
    plt.show()

def plot_sector_comparison(df, year):
    data = df[df['Year'] == year].melt(id_vars=['Year'], var_name='Sector', value_name='Budget')
    fig = px.bar(data, x='Sector', y='Budget', title=f"Budget Allocation in {year}", text='Budget')
    fig.show()

def plot_budget_distribution(df, year):
    data = df[df['Year'] == year].melt(id_vars=['Year'], var_name='Sector', value_name='Budget')
    fig = px.pie(data, names='Sector', values='Budget', title=f"Budget Distribution in {year}")
    fig.show()

if __name__ == "__main__":
    file_path = "Budget_Allocation_India_1990_2024.csv"
    df = load_data(file_path)
    
    plot_budget_trends(df)
    year = int(input("Enter Year for Sector Comparison: "))
    plot_sector_comparison(df, year)
    plot_budget_distribution(df, year)
```
